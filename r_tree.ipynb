```python
from IPython.display import Image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from tqdm import tqdm
%pip install opencv-python
import cv2
import numpy as np
import os
import glob
from rtree import index
```

```python
p = index.Property()
p.dimension = 256

idx = index.Index(properties=p)

# folder_path = "./"  # Replace with the path to your folder
# Replace with the path to your folder
folder_path = 'data/Data/'

# Use glob to get a list of all JPG files in the folder
jpg_files = glob.glob(os.path.join(folder_path, "*.jpg"))
images = {}
features = []
```

```python
# train the model
# Iterate over the JPG files
for i, image_file in tqdm(enumerate(jpg_files)):
    # Load the image
    images[i] = image_file
    image = cv2.imread(image_file)

    # Convert the image from BGR to RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

    # Calculate the color histogram features
    histogram = cv2.calcHist([gray], [0], None, [256], [0, 256])

    # Flatten the histogram into a 1D array
    histogram = histogram.flatten()

    # Normalize the histogram to have values between 0 and 1
    histogram /= np.sum(histogram)

    features.append(histogram)
    idx.insert(i, histogram)
```

```python
# idx.serialize('rtree.dat')
np.save('data/features/', features)
```
